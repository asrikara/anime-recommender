```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("hernan4444/anime-recommendation-database-2020")

print("Path to dataset files:", path)
```

```python
import pandas as pd
```

```python
anime = pd.read_csv(f"{path}/anime.csv")
```

```python
anime
```

```python
synopsis = pd.read_csv(f"{path}/anime_with_synopsis.csv")
```

```python
synopsis = synopsis[['MAL_ID','sypnopsis']]
```

```python
anime_data = pd.merge(anime, synopsis, on='MAL_ID', how='left')
```

```python
anime_data
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
ax = plt.axes()
sns.heatmap(anime_data.isna().transpose(), cbar=False, ax = ax)
#Useless since most of the columns are not missing data
```

```python
anime_data['Genres'].value_counts().reset_index().sort_values("count", ascending=False)
```

```python
anime_data['sypnopsis']
```

```python
anime_data["words_in_syn"] = anime_data["sypnopsis"].str.split().str.len()
```

```python
anime_data
```

```python
anime_data.loc[anime_data['words_in_syn'].between(15, 24), 'sypnopsis']
```

```python
anime_data_20_words = anime_data.loc[anime_data['words_in_syn'] >= 20].copy()
```

```python
anime_data_20_words
```

```python
anime_data_20_words["tagged_syn"] = anime_data_20_words[['MAL_ID', "sypnopsis"]].astype(str).agg(" ".join, axis=1)
```

```python
anime_data_20_words
```

```python
(
    anime_data_20_words.to_csv('anime_cleaned.csv', index = False)
)
```

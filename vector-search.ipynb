```python
from langchain_community.document_loaders import TextLoader
from langchain_text_splitters import CharacterTextSplitter
from langchain_openai import OpenAIEmbeddings
from langchain_chroma import Chroma
```

```python
from dotenv import load_dotenv

load_dotenv()
```

```python
import pandas as pd
```

```python
anime = pd.read_csv('anime_cleaned.csv')
```

```python
anime
```

```python
#The reason for the tagged syn is so that it will make it easier to bring the recommendations rather then having to do a text based match
#Since langchain-text can't use pandas we will have to do something else blow
```

```python
anime['tagged_syn'].to_csv('tagged_syn.txt', sep = '\n', index = False, header = False)
```

```python
raw_documents = TextLoader('tagged_syn.txt', encoding='utf-8').load()
text_splitter = CharacterTextSplitter(chunk_size=0, chunk_overlap=0, separator='\n')
documents = text_splitter.split_documents(raw_documents)
```

```python
db_anime = Chroma.from_documents(
    documents,
    embedding = OpenAIEmbeddings()
)
```

```python
query = "An anime about fighting titans"
docs = db_anime.similarity_search(query, k = 10)
docs
```

```python
anime[anime['MAL_ID'] == int(docs[0].page_content.split()[0].strip())]
```

```python
def retrieve_recommendations(
        query: str,
        top_k: int = 10,
) ->  pd.DataFrame:
    recs = db_anime.similarity_search(query, k = 50)

    anime_list = []

    for i in range (0, len(recs)):
        anime_list += [int(recs[i].page_content.strip('"').split()[0])]

    return anime[anime["MAL_ID"].isin(anime_list)].head(top_k)
```

```python
retrieve_recommendations("An anime about ninjas and naruto")
```

```python
import pandas as pd
```

```python
anime = pd.read_csv('anime_cleaned.csv')
```

```python
anime['Genres'].value_counts().reset_index()
```

```python
# Get all unique genres
all_genres = set()

# Loop through non-null values in the Genre column
for genres_str in anime['Genres'].dropna():
    # Split the string by comma and strip whitespace
    genres = [genre.strip() for genre in genres_str.split(',')]
    # Add these genres to our set
    all_genres.update(genres)

# Convert to sorted list
unique_genres = sorted(list(all_genres))

print(f"Found {len(unique_genres)} unique genres:")
print(unique_genres)
```

```python
anime_filtered = anime[~anime['Genres'].str.contains('Unknown', na=False)]
anime = anime_filtered
```

```python
anime.to_csv('anime_unknown_removed.csv', index=False)
```
